```python
# Slope of linear region = s
V1 = np.array(df1.VBias)
C1 = np.array(df1.C)
# s = []
plt.title("1/C^2 vs V")
plt.xlabel("Vbias")
plt.ylabel("1/C^2")
s1 = np.gradient(1/(C1**2))
plt.plot(V1,1/C1**2)
print(max(s1))
```

```python
C_ox = 1.94*(10**-9)#F
q = 1.6*(10**(-19)) #C
e_s = 12 
e = 8.85*10**(-14)  #F/cm^2
d = 1000*(10**-4)   #cm
A = np.pi*(d**2)/4  #cm^2
kB = 1.38*(10**-23)  
T = 300
# Thickness of oxide
t_ox = (e*e_s*A)/C_ox  #cm
print(f"Thickness of oxide: {t_ox*10**7} nm")

# N_A
slope = 12.125*10**(18) # 1/(F^2*V)
N_D = (2/(q*e_s*e))/(slope*(A**2))
print(f"Doping conceteration : {N_D}/cm^2") #9.83567123326322.1*(10^14)

L_D = np.sqrt((kB*T*e_s*e)/(q**2*N_D))#Diffusion length
print(f"L_D: {L_D}")
C_sFB = e_s*e/L_D
print(f"C_sFB: {C_sFB}")
C_fb = (C_ox*C_sFB)/(C_ox+C_sFB) 

# print(A)
print(f"Flatband capacitance is: {C_fb*10**9} nF")
print(f"Flatband voltage is: 1.1 V")
```
